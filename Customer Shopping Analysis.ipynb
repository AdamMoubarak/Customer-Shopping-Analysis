```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df = df.rename(columns = {'purchase_amount_(usd)': 'purchase_amount'})
```

```python
df.columns
```

```python
# categorizing ages into distinct age groups
age_groups = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q = 4, labels = age_groups)
```

```python
df[['age', 'age_group']].head()
```

```python
frequency_dict = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
df['purchase_frequency_numerical'] = df['frequency_of_purchases'].map(frequency_dict)

    
```

```python
df[['purchase_frequency_numerical', 'frequency_of_purchases']].head()
```

```python
# Is the column discount_applied redundant with promo_code_used?
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# removing redundancy

df = df.drop('promo_code_used', axis = 1)
```
